```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
ds = pd.read_csv('mushrooms.csv')
ds.head()
```

```python
le = LabelEncoder()
df = ds.apply(le.fit_transform)
df.head()
```

```python
data = df.values
data = np.array(data)
np.random.shuffle(data)
print data
```

```python
x = data[:200,1:]
y = data[:200,0]
y[y==0]=-1
print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
def hypo(x,w,b):
    return np.dot(w,x.T)+b

```

```python
def hingeloss(w,b,x,y):
    loss = 0
    loss+=0.5*np.dot(w,w.T)
    
    ti = np.multiply(y,hypo(x,w,b))
    
    ##Eliminating the ON Street points
    ##Allowing Some Error Points
    ti = ti-1
    ti=ti[ti>0]
        
    loss+=np.sum(ti)
    
    return loss
```

```python
def fit(x,y,lr=0.001,n_epochs=500):
    n_features=x.shape[1]
    n_samples = x.shape[0]
    
    w = np.random.random((n_features,))
    b = 0.0
    c = 1.0
    losses = []
    
    for epoch in range(n_epochs):
        loss = hingeloss(w,b,x,y)
        losses.append(loss)
        
        gradb=0.0
        gradw = np.zeros((n_features,))
        
        ti = np.multiply(y,hypo(x,w,b))
        for ix in range(n_features):
            for iy in range(ti.shape[0]):
                if ti[iy]>1:
                    gradw[ix]+=0
                    gradb+=0
                    
                else:
                    gradw[ix]+=c*x[iy][ix]*y[iy]
                    gradb+=c*y[iy]
        w = w -lr*w + lr*gradw
        b = b + lr*gradb
    return w,b

    
```

```python
def acccuracy(x,y,w_fit,b_fit):
    hx = hypo(x,w_fit,b_fit)
    hx[hx>=0]=1
    hx[hx<0]=-1
    print(b_fit)
#     print(hx.shape[0])
    return float(hx[hx==y].shape[0])/y.shape[0]
#     return 0
```

## Fitting and Testing

```python
w_fit,b_fit = fit(x_train,y_train)
```

```python
print(acccuracy(x_test,y_test,w_fit,b_fit))
```
